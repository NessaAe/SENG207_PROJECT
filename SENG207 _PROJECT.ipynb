### **CLEANING THE DATA**

```python
print (df.isnull().sum())
df.dropna(inplace=True)
```

### **HISTOGRAMS OF SELECTED COLUMNS IN THE DATASET**

```python
# HISTOGRAMS

# Plotting Fixed Acidity distribution
df['fixed acidity'].hist(bins=10, color='red')
plt.title('Distribution of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.show()

# Plotting Volatile Acidity distribution
df['volatile acidity'].hist(bins=10, color='orange')
plt.title('Distribution of Volatile Acidity')
plt.xlabel('Volatile Acidity')
plt.ylabel('Frequency')
plt.show()


# Plotting pH distribution
df['pH'].hist(bins=10, color='yellow')
plt.title('Distribution of pH')
plt.xlabel('pH')
plt.ylabel('Frequency')
plt.show()

# Plotting Citric Acid distribution
df['citric acid'].hist(bins=10, color='green')
plt.title('Distribution of Citric Acid')
plt.xlabel('Citric Acid')
plt.ylabel('Frequency')
plt.show()

# Plotting quality distribution
df['quality'].hist(bins=10, color='blue')
plt.title('Distribution of wines by quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()

# Plotting alcohol distribution
df['alcohol'].hist(bins=10, color='indigo')
plt.title('Distribution of Wines by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

### **SCATTER PLOTS SHOWING THE RELATIONSHIPS BETWEEN VARIOUS COLUMNS IN THE DATASET**

```python
#Relationship between volatile acidity and citric acid 
plt.figure(figsize=(10, 6))
plt.scatter(df['volatile acidity'], df['citric acid'], color='brown', alpha=0.6)
plt.title('Scatter Plot: Volatile Acidity vs Citric Acid')
plt.xlabel('Volatile Acidity')
plt.ylabel('citric acid')
plt.grid(True)
plt.show()

#Relationship between residual sugar and sulphates
plt.figure(figsize=(10, 6))
plt.scatter(df['residual sugar'], df['sulphates'], color='cyan', alpha=0.6)
plt.title('Scatter Plot: Residual Sugar vs Sulphates')
plt.xlabel('Residual Sugar')
plt.ylabel('Sulphates')
plt.grid(True)
plt.show()

#Relationship between free sulfur dioxide and total sulfur dioxide
plt.figure(figsize=(10, 6))
plt.scatter(df['free sulfur dioxide'], df['total sulfur dioxide'], color='grey', alpha=0.6)
plt.title('Scatter Plot: Free Sulfur Dioxide vs Total Sulfur Dioxide')
plt.xlabel('Free Sulfur Dioxide')
plt.ylabel('Total Sulfur Dioxide')
plt.grid(True)
plt.show()

#Relationship between alcohol and density
plt.figure(figsize=(10, 6))
plt.scatter(df['alcohol'], df['density'], color='green', alpha=0.6)
plt.title('Scatter Plot: Alcohol vs Density')
plt.xlabel(' Alcohol')
plt.ylabel('Density')
plt.grid(True)
plt.show()


# Relationship between chlorides and quality
plt.figure(figsize=(10, 6))
plt.scatter(df['chlorides'], df['quality'], color='orange', alpha=0.6)
plt.title('Scatter Plot: Chlorides vs Quality')
plt.xlabel('Chlorides')
plt.ylabel('Quality')
plt.grid(True)
plt.show()

# Relationship between fixed acidity and pH
plt.figure(figsize=(10, 6))
plt.scatter(df['fixed acidity'], df['pH'], color='purple', alpha=0.5)
plt.title('Scatter Plot: Fixed acidity vs. pH')
plt.xlabel('Fixed Acidity')
plt.ylabel('pH')
plt.grid(True)
plt.show()



```

### **CORRELATION MAP OF DATASET**

```python
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}", ha='center', va='center', color='w')
        
plt.colorbar()
plt.xticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns)
        
plt.title('Correlation Matrix')
plt.show()
```

## **Correlation Map**
#### The correlation map is represented using the 'coolwarm' color map, where cooler colors (blue) represent negative correlations, warmer colors (red) represent positive correlations, and white represents no correlation (correlation coefficient = 0). The colorbar added on the side of the heatmap, provides a reference for interpreting the colors in the heatmap. It indicates the range of correlation values corresponding to different colors in the plot. The x-axis and y-axis are labeled with the column names from the DataFrame. The rotation=90 argument is used to rotate the x-axis labels by 90 degrees for better readability, especially when dealing with a large number of variables. The color intensity and the numeric annotations aid in interpreting the strength and direction of correlations between different pairs of variables.
